```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

import specrepro as sr
```

```python
# Example usage:
lambdah = 0.0  # Fixed lambda
x = np.linspace(-1, 1, 500)  # Interval [-1, 1] for plotting

# Plot the first 5 Gegenbauer polynomials for fixed lambda
plt.figure(figsize=(8, 6))

for l in range(5):  # Plot for l = 0, 1, 2, 3, 4
    values = sr.gegenbauer_polynomial(l, lambdah, x)
    plt.plot(x, values, label=f'C_{l}^(lambda)(x)')

# Add labels and legend
plt.title(f'First 5 Gegenbauer Polynomials for lambda = {lambdah}')
plt.xlabel('x')
plt.ylabel(f'C_l^(lambda)(x)')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
