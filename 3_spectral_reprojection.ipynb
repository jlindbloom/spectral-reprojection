```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.special import gamma
from scipy.special import jv

import specrepro as sr
```

# $f(x) = x$

```python
L = 1
f = lambda x: x # analytic but not periodic

N = 12
fapprox, fcoeffs = sr.fourier_partial_sum_projection(f, L, N)

m = 2
lambdah = m

# Compute Gegebauer coefficients numerically (not using the Fourier -> Gegenbauer map)
gcoeffs = sr.gegenbauer_coefficients(fapprox, m, lambdah)
gfunc = sr.gegenbauer_partial_sum(gcoeffs, lambdah)

# Plot
dom = np.linspace(-1,1,1000)
plt.figure(figsize=(8,5))
plt.plot(dom, f(dom), label="$f$")
plt.plot(dom, fapprox(dom), label="$S_N f$")
plt.plot(dom, gfunc(dom), label="$P_m^{\\lambda} S_N f$")
plt.title(f"N = {N}, $\\lambda$ = {lambdah}, $m$ = {m} ")
plt.legend()
plt.show()
```

### Vary $m$?

```python
L = 1
f = lambda x: x # analytic but not periodic

N = 12
fapprox, fcoeffs = sr.fourier_partial_sum_projection(f, L, N)

dom = np.linspace(-1,1,1000)
plt.figure(figsize=(8,5))
plt.plot(dom, f(dom), label="$f$")
plt.plot(dom, fapprox(dom), label="$S_N f$")


#ms = [2, 4, 8, 10]
ms = [2, 3, 4, 5, 6, 7]
for m in ms:
    lambdah = m

    # Compute Gegebauer coefficients numerically (not using the Fourier -> Gegenbauer map)
    gcoeffs = sr.gegenbauer_coefficients(fapprox, m, lambdah)
    gfunc = sr.gegenbauer_partial_sum(gcoeffs, lambdah)

    plt.plot(dom, gfunc(dom), label=f"Reproj, m = {m}", ls="--", alpha=0.5)
# plt.title(f"N = {N}, $\\lambda$ = {lambdah}, $m$ = {m} ")
plt.legend()
plt.show()
```
